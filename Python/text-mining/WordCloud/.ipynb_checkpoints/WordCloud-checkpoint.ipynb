```python
with open("./data/abstract.txt", "r", encoding="utf-8") as read:
    abstract = read.readlines()
```

```python
nlp = konlpy.tag.Twitter()
```

```python
tag = Counter(nlp.nouns(str(abstract)))
```

```python
wc = pytagcloud.make_tags(tag.most_common(30), maxsize=100, minsize=10)
```

```python
pytagcloud.create_tag_image(wc, "wc.jpg", fontname='korean', rectangular=True)
```

```python
import nltk
import wordcloud
import konlpy
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
from konlpy.

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
abst = open("./data/abstract.txt", "r", encoding="utf-8").read()
image = np.array(Image.open("./data/alice_mask.png"))
```

```python
t = konlpy.tag.Kkma()
```

```python
t = konlpy.tag.Hannanum()
```

```python
t = konlpy.tag.Twitter()
```

```python
t = konlpy.tag.Komoran()
```

```python
nouns = t.nouns(abst)
```

```python
noun = []
for i in np.arange(len(nouns)):
    temp = nouns.pop()
    if len(temp)>1:
        noun.append(temp)
```

```python
fre = nltk.Text(noun).vocab().most_common(500)
```

```python
abstract = wordcloud.WordCloud(font_path="C:/Windows/Fonts/malgunbd.ttf", 
                               background_color="white", mask=image).generate_from_frequencies(dict(fre))
```

```python
plt.figure(figsize = (15, 15))
plt.imshow(abstract, interpolation="bilinear")
plt.axis("off")
plt.show()
```
